```python
new_columns=[]
for i in laptops.columns:
    a=i.strip()       #removing white spaces.
    new_columns.append(a)
    
laptops.columns=new_columns
```

```python
laptops.head()
```

```python
def clean_col(col):
    col = col.strip()
    col = col.replace("(","")
    col = col.replace(")","")
    col = col.replace("Operating System", "os")
    col = col.replace(" ","_")
    col = col.lower()
    return col

new_columns = []
for c in laptops.columns:
    clean_c = clean_col(c)
    new_columns.append(clean_c)

laptops.columns = new_columns
laptops.columns
```

```python
print(laptops["screen_size"].dtype)
print(laptops["screen_size"].unique())
```

```python
laptops['screen_size']=laptops['screen_size'].str.replace('"','')
```

```python
laptops["ram"] = laptops["ram"].str.replace('GB','')
unique_ram = laptops["ram"].unique()
```

```python
laptops["ram"] = laptops["ram"].str.replace('GB','').astype(int)
laptops.rename({'ram':'ram_gb'},axis=1,inplace=True)
```

```python
#series.str.split()
laptops["gpu_manufacturer"] = (laptops["gpu"]
                                       .str.split()
                                       .str[0]
                              )
laptops['cpu_manufacturer']=laptops['cpu'].str.split().str[0]
cpu_manufacturer_counts=laptops.cpu_manufacturer.value_counts()
```

```python
#series.map()

mapping_dict = {
    'Android': 'Android',
    'Chrome OS': 'Chrome OS',
    'Linux': 'Linux',
    'Mac OS': 'macOS',
    'No OS': 'No OS',
    'Windows': 'Windows',
    'macOS': 'macOS'
}
laptops.os=laptops.os.map(mapping_dict)
```

```python
laptops['price_euros']=laptops['price_euros'].str.replace(',','')
```

```python
laptops.price_euros=laptops.price_euros.astype(float)
```

```python
laptops['resolution']=laptops['screen'].str.split().str[-1] #extracting resolution from screen column
```

```python
laptops['processor_speed']=laptops['cpu'].str.split().str[-1]
```
