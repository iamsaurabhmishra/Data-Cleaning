```python
import pandas as pd
```

```python
billboard=pd.read_csv('/Users/saurabh/cleaning_data/data/billboard.csv')
```

```python
billboard.head(10)
```

```python
billboard_new=pd.melt(billboard,id_vars=['year','artist','track','time','date.entered'],var_name='week')
```

```python
billboard_new.head(10)
```

```python
#now each entry can have repeated entries for many weeks,not just wk1:Billboard logic
#this can be bad for data storage in long term
```

```python
billboard_new.loc[billboard_new['track']=='Try Again'].head(5)  #repeated 76 times
```

```python
billboard_songs=billboard_new[['year','artist','track','time']]   #fixing storage issue by taking only relevant pieces
```

```python
billboard_songs.head()
```

```python
billboard_songs.shape     #still contains duplicates
```

```python
billboard_songs=billboard_songs.drop_duplicates()
```

```python
billboard_songs.shape    #went from 24092 rows to 317:since each song entry was repeated 76 times for 76 weeks.
```
